## Neural Networks
A neural network is a computational model inspired by the way biological neural networks in the human brain operate. It consists of interconnected layers of nodes (or neurons) that process input data to produce an output.

 A neural network typically consists of three types of layers:

  - Input Layer: Accepts the input features.
   - Hidden Layers: Perform computations and feature transformations.
  - Output Layer: Produces the final predictions.

## Working Process of a Neural Network
#### Input Layer:
The network receives input data (features) here.
#### Weights:
Each connection between neurons has an associated weight that determines the importance of the input.
Higher weights mean more influence on the neuron's output.

#### Bias:
Each neuron has a bias, which is an additional parameter added to the weighted sum of inputs. It helps the model make adjustments, allowing it to fit the data better even if all inputs are zero.

#### Summation:
The weighted inputs are summed up, along with the bias. This can be expressed mathematically as:
$$
z = \sum (w_i \cdot x_i) + b
$$
Where:
- \( z \) is the weighted sum (input to the activation function).
- \( w_i \) are the weights.
- \( x_i \) are the input values.
- \( b \) is the bias.


#### Activation Function:
After calculating the weighted sum of inputs plus bias, the result is passed through an activation function (like ReLU or sigmoid) to introduce non-linearity and determine the neuron's output.

#### Output Layer:
The final layer produces predictions based on the outputs from the previous layers.

 - Activation functions determine the output of a neuron and introduce non-linearity into the model.
 - Without activation functions, a neural network would only be able to model linear relationships. By introducing non-linearity, activation functions allow the network to learn and model complex patterns in data. This means the network can solve problems that are not just straight lines, like recognizing images or understanding speech.

Common activation functions include:

#### Sigmoid: 
  - Outputs values between 0 and 1.
   - Useful for binary classification.
   - Example: If the input is 2, the output will be close to 1

#### ReLU (Rectified Linear Unit):
  - Outputs the input directly if its positive; otherwise, it outputs zero.
   -  Introduces sparsity and is computationally efficient.
 - Example: If the input is -3, the output is 0. If the input is 2, the output is 2.
 
#### Tanh
  - Outputs values between -1 and 1
   - Useful for hidden layers as it helps to center the data.

 - Example: If the input is 0.5, the output will be around 0.46.


## Example: Building a Simple Neural Network

```python
import tensorflow as tf
from keras.datasets import mnist
from keras.utils import to_categorical

#Creating a simple neural network to classify handwritten digits from the MNIST dataset.
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X_train = X_train.reshape((60000, 28 * 28)).astype('float32') / 255
#Reshaping 28x28 pixel images (totaling 784 pixels) into a 2D array where each row represents a flattened image.
#There are 60,000 training images.
#/ 255: Normalizes the pixel values from a range of 0-255 (common for image data)
X_test = X_test.reshape((10000, 28 * 28)).astype('float32') / 255
y_train = to_categorical(y_train, 10) #One-hot encoding
y_test = to_categorical(y_test, 10) #one-hot encoding
```

```python
print(X_train)
```

```python
from keras.models import Sequential
from keras.layers import Dense

#Building the Neural Network
#Initializing Sequntial model
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(784,))) # Hidden layer
#Dense(128...) adds a fully connected (dense) layer with 128 neurons.
#input_shape=(784,) defines the shape of the input data. In this case,
#each input is a 1D array of 784 elements (28x28 pixels flattened).
model.add(Dense(10, activation='softmax'))  # Output layer
```

#### Softmax Function:
Converts raw scores into probabilities for multi-class classification
 - Output: Probabilities sum to 1, indicating the likelihood of each class.
  - Usage: Ideal for predicting the class with the highest probability.

For input \( x = [2.0, 1.0, 0.1] \), the softmax output is approximately \( y \approx [0.66, 0.24, 0.10] \).


```python
#Compiling the Model(categorical_crossentropy for categorical classification, adam for optimization)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#Training the Model
model.fit(X_train, y_train, epochs=5, batch_size=32)
```

 - Epochs: Number of complete passes through the training dataset.
 - Batch Size: Number of samples processed before the model's internal parameters are updated.

```python
#Evaluating the Model
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {accuracy}')
```

#### Loss Function:
After predictions are made, the loss function measures the difference between the predicted output and the actual target. It quantifies how well the model is performing.

#### Gradient:
The gradient is calculated based on the loss. It represents the direction and rate of change of the loss function with respect to the weights and biases. This helps in understanding how to adjust the weights and biases to minimize the loss.

#### Backpropagation:
Using the gradients, the network adjusts the weights and biases through a process called backpropagation, effectively learning from the errors and improving over time.



##### Loss measures how wrong the model is.
##### Gradient tells the model how to correct itself to be less wrong in the future. 
##### Backpropagation updates weights and biases to improve the model's accuracy
